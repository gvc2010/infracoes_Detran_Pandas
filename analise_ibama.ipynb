
- **VAL_AUTO_INFRACAO**: Valor do auto de infração
- **UF**: estado onde ocorreu o auto de infração
- **TIPO_AUTO**: tipo da infração
- **DAT_HORA_AUTO_INFRACAO**: data e hora em que a infração foi lavrada

```python
#Importar Dependências

import numpy as np
import pandas as pd
```


Carregar o arquivo "auto_infracao_IBAMA.csv" e armazene em um DataFrame. 
Imprimir as 10 primeiras linhas do DataFrame.

```python
csv = pd.read_csv('auto_infracao_IBAMA.csv') #Ler o arquivo

df = pd.DataFrame(csv) #Transformar em DataFrame

df.head(10) #Printar 10 primeiras linhas
```


Verificar quantos dados faltantes existem na coluna 'VAL_AUTO_INFRACAO' e substituir os dados faltantes por zero. 

```python
faltantes = df['VAL_AUTO_INFRACAO'].isnull().sum() # Verificando os faltantes e somando

print(f'Quantidade de Valores Faltantes:{faltantes}') # Printando faltantes

df['VAL_AUTO_INFRACAO'].fillna(0,inplace=True) #Substituindo null por 0



```

A coluna 'DAT_CIENCIA_AUTUACAO' contém muitos valores faltantes, portanto deve ser eliminada. Eliminar a coluna 'DAT_CIENCIA_AUTUACAO' do DataFrame.

```python
df.drop(columns=['DAT_CIENCIA_AUTUACAO'],inplace=True) #Eliminando Coluna
```

Verificar quantos tipos de infração diferentes exitem na base (coluna 'TIPO_AUTO') e apontar qual o tipo de infração **menos comum** (infração que menos aparece na base). Qual o número de ocorrências da infração menos comum?

```python

print(f"Tipos de Infração: {df['TIPO_AUTO'].nunique()}") #Tipos Únicos de Infração

print(20 * '=')

contagem = df['TIPO_AUTO'].value_counts() #Conta a quantidade
print(f"Contagem de Infrações:\n{contagem}")

print(20 * '=')

# Infração com o menor número de ocorrências
infracao_menos_comum = contagem.idxmin() #Mostra a infração menos comum
quantidade_menos_comum = contagem.min() # Conta a infração menos comum
print(f"Infração com Menores Ocorrências: {infracao_menos_comum} ({quantidade_menos_comum} ocorrências)")
```


Indicar quais estados (coluna 'UF') apresentam mais de 3.000 infrações. Qual o valor total das infrações (soma da coluna 'VAL_AUTO_INFRACAO') no estado com o maior número de infrações? 

```python
# Contar o número de infrações por estado (UF)
infra_por_estado = df['UF'].value_counts() # Conta os estados

# Estados com mais de 3000 infrações
estados_mais_3000 = infra_por_estado[infra_por_estado > 3000] # Se as infrações forem maiores que 3000

# Estado com o maior número de infrações
estado_mais_infracoes = infra_por_estado.idxmax() # Mostra os estados com + de 3000

# Soma total dos valores das infrações para o estado com maior número de infrações
soma_infracoes_estado_max = df[df['UF'] == estado_mais_infracoes]['VAL_AUTO_INFRACAO'].sum()  #Soma para mostrar o estado com mais infrações

# Exibir resultados
print("Estados com mais de 3000 infrações:")
print(estados_mais_3000)

print("\nEstado com mais infrações:", estado_mais_infracoes)
print("Valor total das infrações neste estado:", soma_infracoes_estado_max)

```


Montar o gráfico do número total de infrações aplicadas em cada mês do ano. Especificamente, para cada mês ao longo de todos os anos, some o número de infrações ocorridas naquele mês e faça um gráfico onde o eixo horizontal contém os 12 meses do ano e o vertical o número total de infrações em cada mês.

**Dica**: Gere uma nova coluna no DataFrame com o mês da infração, retirando esta informação da coluna 'DAT_HORA_AUTO_INFRACAO'. Aplique um <font color='blue'> groupby </font> na coluna geranda contando o número de registros em cada grupo. 

```python
import matplotlib.pyplot as plt

#Solucao

df['DAT_HORA_AUTO_INFRACAO'] = pd.to_datetime(df['DAT_HORA_AUTO_INFRACAO'])

df['MES'] = df['DAT_HORA_AUTO_INFRACAO'].dt.month

contagem_infracoes_mensal = df.groupby('MES').size()

plt.figure(figsize=(10,6))
contagem_infracoes_mensal.plot(kind='bar', color='skyblue')
plt.xlabel('Mês')
plt.ylabel('N Infrações')
plt.xticks(ticks=range(0,12), labels=['jan','fev','mar','abr','mai','jun','jul','ago','set','out','nov','dez'])
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()

plt.show
```
